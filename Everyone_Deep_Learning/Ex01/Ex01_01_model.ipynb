```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
import numpy as np 
import tensorflow as tf 
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

#### DataSet
- 0 ~ 17 : attribute 
- 18 : class 

```python
Data_set = np.loadtxt("..\\Data\\ThoraricSurgery.csv", delimiter=",")
Data_set
X = Data_set[:, 0:17]
Y = Data_set[:, 17]
```

#### Sequential 
- DeepLearning Layer Structure 

- 3 layer statistics model < - keras.model.add(3층 모델)
- 2 layer statistics model < - keras.model.add(2층 모델)
- 1 layer statistics model < - keras.model.add(2층 모델)

- 층을 몇 개 쌓을지는 데이터에 따라서 그 때 그때 결정한다. 
- Keras의 장점은 model.add()함수를 사용해서 필여한 만큼의 층을 빠르고 쉽게 쌓을수 있다. 
- Dense( 조밀하게 모여있는 집합 )

- activation Function : 다음 층으로 어떻게 값을 넘길지를 결정하는 부분 여기서는 가장 많이 사용되는 Relu 함수와 Sigmode() 함수를 사용하게 한다. 
- loss 한번의 신경망이 실행될 때 마다 오차값을 추적하는 함수 
- Optimizer : 오차를 어떻게 줄여 나갈지 정하는 함수 


```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

#### loss accuracy
- loss (손실 값)
    - 예측이 실패할 확률
- accuracy (정확도)
    - 예측이 성공할 확률 
    




```python
model.compile(loss="mean_squared_error", optimizer="adam", metrics=["accuracy"])
model.fit(X, Y, epochs=100, batch_size=10)
```
